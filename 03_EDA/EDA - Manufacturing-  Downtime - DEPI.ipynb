```python
columns_to_analyze = ['Temperature', 'Vibration', 'Humidity', 'Pressure', 'Power_Usage']

for col in columns_to_analyze:
    print(f"\nColumn: {col}")
    print("Mean:", df[col].mean())
    print("Max:", df[col].max())
    print("Min:", df[col].min())
```

```python
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1']
plt.figure(figsize=(16,12))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Record_Time'], df['Temperature'], color='red', marker='o', linestyle='-')
plt.title("Temperature Over Time")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Record_Time'], df['Vibration'], color='blue', marker='o', linestyle='-')
plt.title("Vibration Over Time")
plt.xlabel("Time")
plt.ylabel("Vibration")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
colors = ['#FF6F61', '#6B5B95', '#88B04B'] 

df['Machine_Status2'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    figsize=(6,6)
)
plt.title("Machine Status Distribution", color='#6B5B95')
plt.ylabel("")
plt.show()
```

```python
colors = ['#F5DEB3', '#800000', '#FF6347']  #    
colors = ['#F7CAC9', '#92A8D1', '#FFB347']  #   

df['Humidity Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    figsize=(6,6)
)
plt.title("Humidity Status Distribution", color='#92A8D1')
plt.ylabel("")
plt.show()
```

```python
colors = ['#FF6F61', '#77DD77', '#84B6F4', '#FFD700']

plt.figure(figsize=(6,6))
df['Vibration Status'].value_counts().plot(
    kind='pie',
    labels=df['Vibration Status'].value_counts().index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor':'white', 'linewidth':1.5},
    textprops={'fontsize':12, 'weight':'bold'}
)
plt.title("Vibration Status Distribution", color='#FF6F61', fontsize=14, weight='bold')
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,5))
df['RED_Flag'].value_counts().plot(
    kind='pie',
    colors=['#FF6F61','#FFD700'],  #   
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops={'edgecolor':'white', 'linewidth':1.5},
    textprops={'fontsize':12, 'weight':'bold'}
)
plt.title("RED_Flag Distribution", color='#FF6F61', fontsize=14, weight='bold')
plt.ylabel("")
plt.tight_layout()
plt.show()
```
